```python
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
import cv2
import numpy as np
```

```python
from IPython.core.pylabtools import figsize
figsize(10, 10)
```

```python
lena = cv2.imread("lena.png")
lena_gray = cv2.cvtColor(lena, cv2.COLOR_BGR2GRAY)
```

```python
def cv2show(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')
```

```python
cv2show(lena_gray)
```

```python
def prewitt(img : np.ndarray):
    x_derivative = convolve2d(img, [[1, 0, -1]] * 3)
    y_derivative = convolve2d(img, ([[x] * 3 for x in range(1, -2, -1)]))
    return np.sqrt(x_derivative ** 2 + y_derivative ** 2)
```

```python
cv2show(prewitt(lena_gray))
```

```python
cv2show(prewitt(cv2.blur(lena_gray, ksize=(5, 5))))
```
